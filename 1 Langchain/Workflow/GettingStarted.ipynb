```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["LANGCHAIN_API_KEY"] = os.getenv('LANGCHAIN_API_KEY')
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT")
os.environ["LANGCHAIN_TRACING_V2"] = "true"
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o") ## no need to specify the open ai , becuase already importede
print(llm)
```

```python
## Input and get response from LLM
llm.invoke("What is GenAI") ## its paid , otherwise it would have given the answer
```

```python
### ChatPromptTemplate -->its something like mein keh rha hu , mere llm model se ki please behave this way
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages(
           [
               ("system","You are an expert AI engineer.Provide me answers based on the question"),
               ("user","{input}")
           ]     
)
prompt
```

```python
chain = prompt | llm
response = chain.invoke({"input":"Can you tell me about Langsmith?"})  ## its paid , otherwise it would have given the answer
print(response) ## it will give the response along with the content
```

```python
## stroutput Parser
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt | llm | output_parser
response = chain.invoke({"input":"Can you tell me about Langsmith?"})
print(response) ## it will give the response without the content
```
