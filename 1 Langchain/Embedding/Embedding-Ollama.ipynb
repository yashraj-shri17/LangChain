```python
from langchain_community.embeddings import OllamaEmbeddings

embedding = (
                OllamaEmbeddings(model="gemma:2b") ## since we downloaded gemma otherwise bydefault it used llama2
            )
```

```python
embedding
```

```python
r1 = embedding.embed_documents(
                            [
                                "Alpha is the first letter of greek alphabet",
                                "Beta is the second letter of greek alphabet",
                            ]
                                )
```

```python
len(r1[1])
```

```python
r1[1]
```

```python
### Querying
embedding.embed_query("What is the second letter of Greek Alphabet")
```

```python
#3 Training One More Model
embedding2 = OllamaEmbeddings(model="mxbai-embed-large")
text = "I enjoy hiking in the mountains under a clear blue sky."
query_result = embedding2.embed_query(text)
query_result,len(query_result)
```

```python
query_result2 = embedding2.embed_query("The stock market crashed causing widespread economic panic.")
```

```python
## comparing similarity
import numpy as np
v1 = np.array(query_result)
v2 = np.array(query_result2)

cos_sim = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
print("Cosine similarity between the two embeddings:", cos_sim)
```
